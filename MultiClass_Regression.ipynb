```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
data=pd.read_csv('Iris.csv')
data
```

```python
data['Species'].unique()
```

```python
data['Species'].replace({'Iris-setosa':1,'Iris-versicolor':2,'Iris-virginica':3})
```

```python
data
```

```python
X_train, X_test, y_train, y_test = train_test_split(data[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']],data['Species'], test_size=0.3) # Corrected column names to match the DataFrame
```

```python
X_train
```

```python
y_train
```

```python
from sklearn import linear_model
mymodel = linear_model.LogisticRegression(max_iter=120)
mymodel.fit(X_train,y_train)
```

```python
from sklearn.linear_model import LogisticRegression # make sure to import the class

LogisticRegression(C=1.0, class_weight=None, dual=False, fit_intercept=True,
                   intercept_scaling=1, l1_ratio=None, max_iter=120,
                   multi_class='auto', n_jobs=None, penalty='l2',
                   random_state=None, solver='lbfgs', tol=0.0001, verbose=0,
                   warm_start=False)
```

```python
mymodel.predict(X_test)
```

```python
X_test
```

```python
mymodel.score(X_test, y_test)
```

```python
predicted_output = mymodel.predict(X_test)
predicted_output
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, predicted_output)
cm
```

```python
import seaborn as sn
from matplotlib import pyplot as plt
plt.figure(figsize = (5,4))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted Value')
plt.ylabel('Truth or Actual Value')
```
